# Netflix Movie Recommendation System using Apriori Algorithm

## Project Summary
This project implements the Apriori algorithm to discover association rules in a Netflix movie dataset. The goal is to identify movie pairings that frequently appear together, which can be used to build a recommendation system. By analyzing movie purchases or viewing histories, we can suggest other movies to users based on their preferences, enhancing user experience and engagement on the platform.

## Project Introduction
This project aims to leverage the power of association rule mining, specifically the Apriori algorithm, to develop a movie recommendation system for Netflix. By analyzing a dataset of movie viewing patterns, we can uncover implicit relationships between movies. Understanding these relationships allows us to recommend movies that users are likely to enjoy based on their past viewing history, thereby improving user satisfaction and increasing watch time on the platform.

## Methodology
The core of this recommendation system is the Apriori algorithm. The methodology involves the following steps:

1.  **Data Loading and Preprocessing**: The Netflix movie dataset is loaded, and raw viewing data is transformed into a transactional format suitable for association rule mining. Each transaction represents a user's movie selection, where each item in the transaction is a movie.
2.  **Apriori Algorithm Application**: The Apriori algorithm is applied to the processed transactional data to find frequent itemsets and generate strong association rules. Key parameters like `min_support`, `min_confidence`, and `min_lift` are configured to filter out less significant rules.
3.  **Rule Interpretation**: The generated association rules are then inspected and transformed into a human-readable format, showing which movies tend to be watched together.
4.  **Recommendation Generation**: The extracted rules form the basis of the recommendation system. When a user watches a movie, the system can look for rules where that movie is in the antecedent and suggest movies from the consequent.
5.  **Visualization**: The top association rules are visualized to provide a clear understanding of the strongest movie pairings found in the dataset.

## Why Apriori?
The Apriori algorithm was selected for this project due to its foundational role in association rule mining and its effectiveness in identifying frequent itemsets from transactional datasets. Here are the key reasons for its choice:

*   **Simplicity and Interpretability**: Apriori is relatively straightforward to understand and implement. The resulting association rules are easy to interpret, making it simple to explain why certain recommendations are made.
*   **Well-suited for Transactional Data**: The Netflix movie viewing history, when framed as transactions (movies watched by a user), aligns perfectly with the type of data Apriori is designed to process.
*   **Identification of Strong Associations**: By setting appropriate `min_support`, `min_confidence`, and `min_lift` thresholds, Apriori can effectively filter out spurious relationships and highlight genuinely strong and interesting associations between movies.
*   **Scalability for Sparse Data**: While its performance can degrade with very dense datasets or extremely low support thresholds, for many recommendation scenarios with moderately sparse data, Apriori provides a good balance of performance and result quality.
*   **Foundation for Further Exploration**: Apriori provides a solid baseline for understanding movie co-occurrence. Its output can serve as a valuable input for more complex recommendation engines or personalized filtering methods.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dataset = pd.read_csv('/content/Netflix_Movie_Recommendation.xlsx - Netflix Movie Recommendation Ec.csv',header=None)
transactions =[]
for i in range(0,7466):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

## 2. Data Loading and Preprocessing

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

## 3. Applying the Apriori Algorithm

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
  movie_1 = [tuple(result[2][0][0])[0] for result in results]
  movie_2 = [tuple(result[2][0][1])[0] for result in results]
  supports=[result[1] for result in results]
  return list(zip(movie_1,movie_2,supports))
resultsinDataFrame=pd.DataFrame(inspect(results),columns=['Movie 1','Movie 2','Support'])
```

## 4. Inspecting and Formatting Results

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n=10,columns='Support')
```

## 5. Visualizing Top Association Rules

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 association rules by support
top_rules = resultsinDataFrame.nlargest(n=10, columns='Support')

# Create a combined 'Association' column for better labeling on the plot
top_rules['Association'] = top_rules['Movie 1'] + ' -> ' + top_rules['Movie 2']

# Plotting the top association rules
plt.figure(figsize=(12, 8))
sns.barplot(x='Support', y='Association', data=top_rules, palette='viridis', hue='Association', legend=False)
plt.title('Top 10 Movie Association Rules by Support', fontsize=16)
plt.xlabel('Support', fontsize=12)
plt.ylabel('Movie Association', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
